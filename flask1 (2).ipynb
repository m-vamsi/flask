```python
import numpy as np
import pandas as pd
from flask_table import Table, Col

ds=pd.read_csv("final_data.csv")
ds
```

```python




class ItemTable(Table):
    sno=Col('S')
    cid=Col('CustomerID')
    gen= Col('Gender')
    age= Col('Age')
    anIn=Col('Annual.Income..k..')
    spsc=Col('Spending.Score..1.100')
    seg=Col('segments')

class Item(object):
    def __init__(self,sno,cid,gen,age,anIn,spsc,seg):
        self.sno=ds.iloc[:,0].values
        self.cid=ds.iloc[:,1].values
        self.gen=ds.iloc[:,2].values
        self.age=ds.iloc[:,3].values
        self.anIn=ds.iloc[:,4].values
        self.spsc=ds.iloc[:,5].values
        self.seg=ds.iloc[:,6].values
items=[]
for i in range(1,201):
    items.append(Item('sno[i]','cid[i]','gen[i]','age[i]','anIn[i]','spsc[i]','seg[i]'))
table = ItemTable(items) 
print(table.__html__())
```
